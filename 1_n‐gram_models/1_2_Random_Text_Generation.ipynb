# 1.2 Random Text Generation

```python
from utils1 import load_dataset, simple_tokenize
from ngram import BasicNgram
```

```python
text = load_dataset("The Jungle Book")
tokenized_text = simple_tokenize(text, remove_symbols=True)
print(tokenized_text[4200:4300])
```

```python
LENGTH = 100
START = ["in", "the", "end"]

for _ in range(4):
    for n in range(1, len(START)+1):
        print(f"N={n+1}", end=" ")
        results = list(START[:n])
        ngram = BasicNgram(n+1, tokenized_text)
        for _ in range(LENGTH):
            results.append(ngram[tuple(results[len(results)-n:])].generate())
        print(" ".join(results))
    print()
```

## Discussion

- When N=1, the generated text is hard to understand.
- As N increase, the sentences become more meaningful.
- When N increase larger, the model may raise an exception. Because it has generated a context that never occurs in the training set, and therefore unable to report the distribution of next word.
